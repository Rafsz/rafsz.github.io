#### - We will use Pandas to download the quotes for two stocks traded in the Brazilian Market (Bovespa/B3) RAPT4 and RAPT3 since the beginning of dec/2020 (12months)
#### - Use Matplotlib to plot the graph for these stocks
#### - Compare their performances
#### - Check Dividends paid in this period for each stock
#### - Get to a conclusion about which stock was a better investment in the past 12 months

```python
import pandas as pd
from pandas_datareader import data as web
```

```python
import matplotlib.pyplot as plt
```

# 1. Getting quotes for RAPT4

```python
dfrapt4 = web.DataReader('RAPT4.SA', data_source="yahoo", start = '12-01-2020', end = '11-26-2021')
display (dfrapt4)
```

```python
dfrapt4["Adj Close"].plot(figsize=(15,8))
plt.legend()
plt.grid();
```

# 2. Getting quotes for RAPT3

```python
dfrapt3 = web.DataReader('RAPT3.SA', data_source="yahoo", start = '12-01-2020', end = '11-26-2021')
display (dfrapt3)
```

```python
dfrapt3["Adj Close"].plot(figsize=(15,8))
plt.legend()
plt.grid();
```

# 3. RAPT4 vs RAPT3

```python
dfrapt4["Adj Close"].plot(figsize=(20,5), label="RAPT4")
dfrapt3["Adj Close"].plot(figsize=(20,5), label="RAPT3")
plt.legend()
plt.grid();
```

#####  (obs: for exact values, check the table/df)
#### *In the past 12months RAPT4 dropped from R$14 to R$11, that is: -21.5% .*
#### *While RAPT3 went from R$10.50 to 13, that is: +24%.*
#### *Maybe the reason is the Dividends Paid for each stock?*

## Considering both stocks paid the same amount of dividends in the past 12 months, RAPT3 was by far a better investment than RAPT4!
#### *Why?...*
